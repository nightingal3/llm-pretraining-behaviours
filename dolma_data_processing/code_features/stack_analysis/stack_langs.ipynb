```python
import os
import pandas as pd
import pyarrow.parquet as pq
import random
from tqdm import tqdm
```

```python
path = "/data/tir/projects/tir7/user_data/mchen5/dolma_1T_deduped/stack-code"
files = os.listdir(path)
langs = {}
```

```python
for f in tqdm(files):
    df = pq.read_table(os.path.join(path, f)).to_pandas()
    for lang in df["lang"]:
        if lang in langs:
            langs[lang] += 1
        else:
            langs[lang] = 1
```

```python
langs = dict(sorted(langs.items(), key=lambda item: item[1], reverse=True))
langs_df = pd.DataFrame(list(langs.items()), columns=["Language", "Count"])
langs_df["Percent"] = langs_df["Count"] / langs_df["Count"].sum() * 100
langs_df["Percent prefix sum"] = langs_df["Percent"].cumsum()
pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)
pd.set_option("display.width", None)
pd.set_option("display.max_colwidth", 0)
langs_df
```

```python
fake_langs = [
    "html",
    "css",
    "scss",
    "markdown",
    "text",
    "xml",
    "yaml",
    "groff",
    "asciidoc",
    "gettext-catalog",
    "turtle",
    "linker-script",
    "qmake",
    "textile",
    "web-ontology-language",
]
langs_df_real = langs_df[~langs_df["Language"].isin(fake_langs)].reset_index(drop=True)
langs_df_real["Percent"] = langs_df_real["Count"] / langs_df_real["Count"].sum() * 100
langs_df_real["Percent prefix sum"] = langs_df_real["Percent"].cumsum()
langs_df_real
```

```python
langs_df.to_csv("stack_breakdown.csv", index=False)
langs_df_real.to_csv("stack_breakdown_real.csv", index=False)
```
