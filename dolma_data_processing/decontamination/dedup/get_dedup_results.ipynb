```python
import os
import sys
import pandas as pd
import pyarrow.parquet as pq
from tqdm import tqdm

domains = ["c4", "common-crawl", "gutenberg-books", "peS2o", "stack-code", "wiki-en-simple"]
DEDUPED = "/data/tir/projects/tir7/user_data/mchen5/dolma_1T_deduped"
```

```python
for domain in domains:
    dir = os.path.join(DEDUPED, domain)
    print(domain)
    total = 0
    dirty = 0
    for filename in tqdm(os.listdir(dir)):
        df = pq.read_table(os.path.join(dir, filename)).to_pandas()
        dirty += len(df[(df["num_contaminated"] != 0)])
        total += len(df)
    print(f"{dirty} of {total} lines contamianted")
    dirty_percent = (dirty / total) * 100
    print(f"{dirty_percent:.2f}% of the lines are contaminated.")
```
