```python
with open('./arithmetic.txt', 'r') as file:
    arithmetic = file.read()
```

```python
# Test decontaminating arithmetic against itself
# Entire string should be contaminated, so result should be empty
data = arithmetic
contaminant = arithmetic

# TODO: Figure out if we should delete special characters
# If we delete special characters from the tasks, we should probably delete them from the data too
janitor = Janitor(delete_chars = "")
janitor.register_contaminant(contaminant)

result = janitor.clean_python(data)
result = "".join(result)

assert result == ''

janitor.dirt_ngrams
```

```python
# Test decontaminating arithmetic against a bogus string
# Nothing should be contaminated, so the result should be the original data
data = arithmetic
contaminant = "Bing chilling!"

janitor = Janitor(delete_chars = "")
janitor.register_contaminant(contaminant)

result = janitor.clean_python(data)
result = "".join(result)

result == data
```
